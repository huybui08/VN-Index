```python
import pandas as pd

# Bước 1: Đọc dữ liệu từ file CSV (nếu chưa làm)
file_path = 'GDP.csv'  # Thay thế đường dẫn của bạn
df = pd.read_csv(file_path)

# Bước 2: Xóa cột 'Đơn vị tính'
df = df.drop(columns=['Đơn vị tính'])

# Bước 3: Tính phần trăm giá trị của mỗi hàng so với hàng trước đó
df['% Value'] = df['Value']  # Tạo cột mới '% Value'
for i in range(1, len(df)):
    df['% Value'].iloc[i] = (df['Value'].iloc[i] * 100) / df['Value'].iloc[i - 1]

# Bước 4: Gán giá trị của hàng đầu tiên thành 100% (vì không có giá trị tháng trước để so sánh)
df['% Value'].iloc[0] = 100

# Bước 5: Làm tròn giá trị '% Value' đến 2 chữ số thập phân
df['% Value'] = df['% Value'].round(2)
df = df.drop(columns=['Value'])
df = df.rename(columns={'% Value':'Value'})
# Bước 6: Hiển thị kết quả
print(df)

# Bước 7: Lưu lại nếu cần
df.to_csv('GDP(%).csv', index=False)
```
